```python
# exercise 4
# implementing the Counting Ones problem using (1+1)-GA
# goal: find a bit string whose sum of its entries is maximum so our goal is to have
#       a string of 100 ones
get_ipython().magic('matplotlib inline')
import random
import numpy as np
from random import randint
import matplotlib.pyplot as plt 
```

```python
# given parameters
l=100
p_m=1/l
```

```python
def mutation(x,p_m):
    gene=x
    for i in range(len(gene)):
        r = random.random()
        if (r < p_m):
            gene[i] = not gene[i]
    return gene

def compute_fitness(array):
    return np.sum(array)

# x is our inital bit sequence randomly generated 
def create_initial(l):
    x= np.random.randint(2,size=(l,))
    return x
```

```python
def genetic_alg(length,probability_mutation,iterations):
    #  create an array to memorize fitness
    fitness_all=[]
    
    
    #a) Randomly generate a bit sequence x
    x = create_initial(length)
    fitness_x= compute_fitness(x)
    fitness_all.append(fitness_x)
    
    #d) repeat
    for i in range(0,iterations):
        
        #b)  Create a copy of x and invert each of its bits with probability p. Let xm be the result.
        copy_x = copy.deepcopy(x)
        xm = mutation(copy_x,1/length)
        fitness_xm = compute_fitness(xm)
        
        #c)If xm is closer to the goal sequence than x then replace x with xm.
        if fitness_xm >= fitness_x:
            x=xm
            fitness_x=fitness_xm
            
        fitness_all.append(fitness_x)
    
    axes = plt.gca()
    axes.set_xlim([0,iterations])
    axes.set_ylim([0,100])    
    plt.plot(fitness_all)
    plt.xlabel('iterations')
    plt.ylabel('fitness')
    plt.title('(1+1)-GA')

    plt.show()
    return fitness_all
```
